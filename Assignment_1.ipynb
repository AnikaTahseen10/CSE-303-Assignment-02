```python
# Calculate the frequency of each category in 'Air Quality'
category_counts = df['Air Quality'].value_counts()

# Print the counts and percentages
print(category_counts)
print("\nPercentage Distribution:")
print(category_counts / len(df) * 100)
```

 If the categories in 'Air Quality' have roughly equal counts or percentages, then the dataset is considered balanced.<br>
 If one or more categories significantly outnumber others, the dataset is imbalanced.<br>
 In this example, the dataset is imbalanced because the 'Good' category has a significantly higher representation than the other categories.<br>
 Generally, highly imbalanced datasets can lead to biased models that perform poorly on minority classes.

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = df.select_dtypes(include=np.number).columns

for column in df.columns:
    if column in numerical_cols and df[column].nunique() > 10:  # continuous numerical data
        # mean, median, variance, and standard deviation
        mean = df[column].mean()
        median = df[column].median()
        variance = df[column].var()
        std_dev = df[column].std()

        print(f"Statistics for {column}:")
        print(f"Mean: {mean}")
        print(f"Median: {median}")
        print(f"Variance: {variance}")
        print(f"Standard Deviation: {std_dev}")

        # histogram
        plt.figure(figsize=(8, 6))
        sns.histplot(df[column], kde=True, bins=30,  color='#0d88e6')
        plt.title(f"Distribution of {column}")
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.show()

    else:  # Categorical
        value_counts = df[column].value_counts()
        percentages = value_counts / len(df) * 100

        print(f"Frequency and Percentage for {column}:")
        print(value_counts)
        print("\nPercentages:")
        print(percentages)

        fig, axes = plt.subplots(1, 2, figsize=(18, 8))  # two subplots

        # Bar chart
        sns.countplot(x=column, data=df, ax=axes[0], color='#fd7f6f')
        axes[0].set_title(f"Bar Chart of {column}")
        axes[0].set_xlabel(column)
        axes[0].set_ylabel("Frequency")
        plt.xticks(rotation=45, ha='right')

        # Pie chart
        axes[1].pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=90)
        axes[1].set_title(f"Pie Chart of {column}")

        plt.tight_layout()
        plt.show()
```

7.

```python
# Calculate the correlation matrix
numerical_df = df.select_dtypes(include=np.number)
correlation_matrix = numerical_df.corr()

# Create the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Pollution Dataset')
plt.show()
```

We use df.isnull().sum() to check for missing values in each column and get the count of missing values.<br>
a) Removing rows with missing values:<br>
 Method: Use df.dropna() to remove rows containing any missing values.<br>
Reasoning: This is a simple approach, but it can lead to loss of information if there are many missing values or if the missing values are not randomly distributed.<br>
b) Imputing missing values:<br>

Method: Use imputation techniques to fill in missing values with estimated values. Some common imputation methods include:<br>
Mean/Median imputation: Replace missing values with the mean or median of the column.<br>
K-Nearest Neighbors (KNN) imputation: Use the values of nearest neighbors to estimate missing values.<br>
Regression imputation: Use a regression model to predict missing values based on other columns.<br>
Reasoning: Imputation can help preserve data and avoid bias, but it's important to choose an appropriate imputation method based on the data characteristics.
